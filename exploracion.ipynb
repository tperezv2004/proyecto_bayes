# Base de datos

```python
df = pd.read_csv("Data\completos.csv")
```

# Filtrar solo chile

```python
# Filtrar competencias realizadas en Chile

df_chile = df[
    df["Country"]
    .astype(str)
    .str.strip()
    .str.lower()
    .eq("chile")
].copy()

print(f"Filas totales: {len(df):,}")
print(f"Filas solo Chile: {len(df_chile):,}")

df_chile.head()
```

# Guardar datos

```python
df_chile.to_csv("Data\solo_chile.csv", index=False)
```
